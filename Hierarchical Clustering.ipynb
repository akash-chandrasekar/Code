```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
uni_df = pd.read_excel('Universities.xlsx')
```

```python
uni_df
```

```python
# Normalising the data except Univ column
def norm_fun(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# removing column Univ using iloc

norm_df = norm_fun(uni_df.iloc[:,1:])
```

```python
norm_df
```

```python
# Now we construct dendogram
dendogram = sch.dendrogram(sch.linkage(norm_df))
```

```python
hcl = AgglomerativeClustering(n_clusters=5,metric='euclidean')
```

```python
y = hcl.fit_predict(norm_df)
```

```python
y
```

```python
uni_df['clusters'] = y
```
