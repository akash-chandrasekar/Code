```python
le = LabelEncoder()
```

```python
card_df['SEX'] = le.fit_transform(card_df['SEX'])
card_df['EDUCATION'] = le.fit_transform(card_df['EDUCATION'])
card_df['MARRIAGE'] = le.fit_transform(card_df['MARRIAGE'])
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x = card_df.drop(labels= 'DEFAULT',axis =1)
```

```python
y = card_df['DEFAULT']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
rf = RandomForestClassifier()
adaboost = AdaBoostClassifier()
gradient_boosting = GradientBoostingClassifier()
xgboost = XGBClassifier()
lbgm = LGBMClassifier()
```

```python
%%time
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
accuracy_score(y_test,y_pred_rf)
```

```python
%%time
adaboost.fit(x_train,y_train)
```

```python
y_pred_adaboost = adaboost.predict(x_test)
accuracy_score(y_test,y_pred_adaboost)
```

```python
%%time
gradient_boosting.fit(x_train,y_train)
```

```python
y_pred_gradient_boosting = gradient_boosting.predict(x_test)
accuracy_score(y_test,y_pred_gradient_boosting)
```

```python
%%time
xgboost.fit(x_train,y_train)
```

```python
y_pred_xgboost = xgboost.predict(x_test)
accuracy_score(y_test,y_pred_xgboost)
```

```python
%%time
lbgm.fit(x_train,y_train)
```

```python
y_pred_lbgm = lbgm.predict(x_test)
accuracy_score(y_test,y_pred_lbgm)
```

## feature importance

```python
rf_feature = rf.feature_importances_
adaboost_feature = adaboost.feature_importances_
gradient_feature = gradient_boosting.feature_importances_
xgboost_feature = xgboost.feature_importances_
lbgm_feature = lbgm.feature_importances_
```

```python
feature_importance = pd.DataFrame({'Features':x_train.columns,
                                  'rf_feature': rf_feature,
                                  'adaboost_feature':adaboost_feature,
                                  'gradient_feature':gradient_feature,
                                  'xgboost_feature':xgboost_feature,
                                  'lbgm_feature': lbgm_feature})
```

```python
feature_importance
```

```python
import seaborn as sns

adaboost_imp = feature_importance.sort_values(by = 'adaboost_feature',ascending=False)
sns.barplot(x = 'adaboost_feature', y = 'Features', data = adaboost_imp )
```

```python
rf_imp = feature_importance.sort_values(by = 'rf_feature',ascending=False)
sns.barplot(x = 'rf_feature', y = 'Features', data = rf_imp,)
```
