```python
# creating pivot table

france_pivot = pd.pivot_table(data = france_df, values= 'Quantity',columns='Description',index = 'InvoiceNo').fillna(0)
```

```python
france_pivot
```

```python
# replaceing values which is more than 1 using lamba function
france_encoded = france_pivot.applymap(lambda x:1 if x>0 else 0)
```

```python
france_encoded
```

```python
frequent_itemset = apriori(france_encoded,
    min_support=0.05,
    use_colnames=True)
```

```python
best_association = association_rules(frequent_itemset)
```

```python
best_association
```
