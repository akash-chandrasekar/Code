```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
x = pd.DataFrame(data = data.data,columns=data.feature_names)
```

```python
x
```

```python
y = data.target
```

```python
y
```

```python
x.shape
```

```python
x.info()
```

```python
x.isna().sum()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.feature_selection import SelectFromModel 
```

```python
model_select = SelectFromModel(estimator=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=5))
```

```python
model_select.fit(x_train,y_train)
```

```python
# It will return which of columns are important each time we run no. of columns get varied
x.columns[model_select.get_support()]
```

```python
x_train_transform = model_select.transform(x_train)
```

```python
x_test_transform = model_select.transform(x_test)
```

```python
x_train_transform.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test_transform.shape
```

```python
def run_randomforest(x_train,x_test,y_train,y_test):
    rf_model = RandomForestClassifier(max_depth=5)
    rf_model.fit(x_train,y_train)
    y_pred = rf_model.predict(x_test)
    print('Accuracy score is:',accuracy_score(y_test,y_pred))
    print('Confusion matrix is:',confusion_matrix(y_test,y_pred))
```

```python
%%time
run_randomforest(x_train,x_test,y_train,y_test)
```

```python
%%time
run_randomforest(x_train_transform,x_test_transform,y_train,y_test)
```

## The above approach we can't able to control the feature for every run the no. of feature will get vary to overcome that we have RFE
## recursive feature elimination

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(estimator=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=5))
```

```python
rfe.fit(x_train,y_train)
```

```python
x.columns[rfe.get_support()]
```

```python
x_train_transform_rfe = rfe.transform(x_train)
x_test_transform_rfe = rfe.transform(x_test)
```

```python
%%time
run_randomforest(x_train_transform_rfe,x_test_transform_rfe,y_train,y_test)
```

```python
# selecting n no. of features using rfe with for loop

for i in range(1,len(x.columns)):
    rfe = RFE(estimator=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=10),n_features_to_select=i)
    rfe.fit(x_train,y_train)

    x_train_transform_rfe = rfe.transform(x_train)
    x_test_transform_rfe = rfe.transform(x_test)

    
    print(x_train_transform_rfe.shape)

    print('Number of features',i)

    run_randomforest(x_train_transform_rfe,x_test_transform_rfe,y_train,y_test)

    print('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
    
```
