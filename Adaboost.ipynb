# why stratify

## To bring the data to similar form

```python
# using stratify

x_train,x_pred,y_train,y_pred = train_test_split(x,y,train_size = 0.8,random_state=24,stratify=y)
```

```python
plt.pie(y.value_counts(),autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(y_train.value_counts(),autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(y_pred.value_counts(),autopct = '%0.1f%%')
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
# In default adaboost takes decision tree
model_ada = AdaBoostClassifier(n_estimators=100)
```

```python
model_ada.fit(x_train,y_train)
```

```python
y_train_pred = model_ada.predict(x_train)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
y_test_pred = model_ada.predict(x_pred)
```

```python
accuracy_score(y_pred,y_test_pred)
```
