## Conclusion:

## if the user watching 'Dangal' we can recommend 'Mary Kom' and 'The legend of Bhagat Singh'

# now we create a function to recommend the movie for the user


```python
def movie_rec(name,n):
    linear_value = linear_kernel(genre_tranform[indexed[name]],genre_tranform)
    scores=pd.DataFrame(linear_value).T
    scores.columns=["score"]
    score = scores.sort_values(by = 'score',ascending=False)

    list_movie = []
    for i in range(0,n+1):
        s = {'Movie name': movie['Movie Title'][score.index[i]],
             'Ratinng': movie['Ratinng'][score.index[i]],
             'genre': movie['genre'][score.index[i]],
             'act': movie['act'][score.index[i]]
             }
        list_movie.append(s)
    return list_movie
    
```

```python
user1 = pd.DataFrame(movie_rec('Dangal',10))
user1
```

```python
user2 = pd.DataFrame(movie_rec('Taare Zameen Par',10))
user2
```
